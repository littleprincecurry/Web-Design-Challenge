```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.pyplot import figure
import numpy as np
```

```python
from jikanpy import Jikan
jikan = Jikan()
```

```python
watched_page_one = jikan.user('scotwilson', 'animelist', 'completed')
watched_page_one
```

```python
watched_page_two = jikan.user('scotwilson', 'animelist', 'completed', page=2)
watched_page_two
```

```python
watched_page_three = jikan.user('scotwilson', 'animelist', 'completed', page=3)
watched_page_three
```

```python
df = pd.DataFrame()
df['title'] = ''
df['mal id'] = ''
df['type'] = ''
df['score'] = ''
df['total episodes'] = ''
```

```python
id_list = []
anime_list = []
type_list = []
score_list = []
total_ep_list = []

for anime in watched_page_one['anime']:
    id_list.append(anime['mal_id'])
    anime_list.append(anime['title'])
    type_list.append(anime['type'])
    score_list.append(anime['score'])
    total_ep_list.append(anime['total_episodes'])
    
for anime in watched_page_two['anime']:
    id_list.append(anime['mal_id'])
    anime_list.append(anime['title'])
    type_list.append(anime['type'])
    score_list.append(anime['score'])
    total_ep_list.append(anime['total_episodes'])
    
for anime in watched_page_three['anime']:
    id_list.append(anime['mal_id'])
    anime_list.append(anime['title'])
    type_list.append(anime['type'])
    score_list.append(anime['score'])
    total_ep_list.append(anime['total_episodes'])
    

```

```python
df['mal id'] = id_list
df['title'] = anime_list
df['type'] = type_list
df['score'] = score_list
df['total episodes'] = total_ep_list
```

```python
# df = df.set_index('mal id')
```

```python
df.head()
```

```python
for show in id_list:
    print(jikan.anime(show))
    time.sleep(2)
    
```

```python
df['english title'] = ''
df['air date start'] = ''
df['air date end'] = ''
df['average score'] = ''
df['rank'] = ''
df['popularity'] = ''
df['genre'] = ''
df['studio'] = ''
df['source'] = ''
df['season'] = ''
```

```python
jikan.anime(306)['studios']
```

```python
for index, row in df.iterrows():
    id = row['mal id']
    response = jikan.anime(id)
    time.sleep(5)
    try:
        df.loc[index, 'english title'] = response['title_english']
        df.loc[index, 'source'] = response['source']
        df.loc[index, 'air date start'] = response['aired']['from']
        df.loc[index, 'air date end'] = response['aired']['to']
        df.loc[index, 'average score'] = response['score']
        df.loc[index, 'rank'] = response['rank']
        df.loc[index, 'popularity'] = response['popularity']
        df.loc[index, 'genre'] = response['genres'][0]['name']
        df.loc[index, 'studio'] = response['studios'][0]['name']
        df.loc[index, 'source'] = response['source']
        df.loc[index, 'season'] = response['premiered']
    except:
        print('whoops')
        
```

```python
df.to_csv('animelist.csv')
```
