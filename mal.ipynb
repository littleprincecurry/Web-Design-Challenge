```python
import pandas as pd
```

```python
from jikanpy import Jikan
jikan = Jikan()
```

```python
watched_page_one = jikan.user('scotwilson', 'animelist', 'completed')
watched_page_one
```

```python
watched_page_two = jikan.user('scotwilson', 'animelist', 'completed', page=2)
watched_page_two
```

```python
watched_page_three = jikan.user('scotwilson', 'animelist', 'completed', page=3)
watched_page_three
```

```python
df = pd.DataFrame()
df['title'] = ''
df['mal id'] = ''
df['type'] = ''
df['score'] = ''
df['total episodes'] = ''
```

```python
id_list = []
anime_list = []
type_list = []
score_list = []
total_ep_list = []

for anime in watched_page_one['anime']:
    id_list.append(anime['mal_id'])
    anime_list.append(anime['title'])
    type_list.append(anime['type'])
    score_list.append(anime['score'])
    total_ep_list.append(anime['total_episodes'])
    
for anime in watched_page_two['anime']:
    id_list.append(anime['mal_id'])
    anime_list.append(anime['title'])
    type_list.append(anime['type'])
    score_list.append(anime['score'])
    total_ep_list.append(anime['total_episodes'])
    
for anime in watched_page_three['anime']:
    id_list.append(anime['mal_id'])
    anime_list.append(anime['title'])
    type_list.append(anime['type'])
    score_list.append(anime['score'])
    total_ep_list.append(anime['total_episodes'])
    

```

```python
df['mal id'] = id_list
df['title'] = anime_list
df['type'] = type_list
df['score'] = score_list
df['total episodes'] = total_ep_list
```

```python
df = df.set_index('mal id')
```
